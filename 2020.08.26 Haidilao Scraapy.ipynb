```python
# scrape for all cities and store as json files in hourly buckets
import requests
import time
from datetime import datetime
import os

dateTimeObj = datetime.utcnow()
dateBucket = dateTimeObj.strftime("%Y-%m-%d")
hourBucket = dateTimeObj.strftime("%H")
path = './haidilao/%s/%s/' % (dateBucket, hourBucket)
os.makedirs(path, exist_ok=True)

url = 'https://superapp.kiwa-tech.com/app/v2/getStoreList'
for city in cities['cityId']:
    try:
        print('requesting %s store list...' % city)
        params = {"customerId": "",
                  "coordinate": "114.217261, 22.380012",
                  "cityId": city,
                  "regionId": "",
                  "sort": ""
                 }

        r = requests.post(url, json = params)
        outfile = './haidilao/%s/%s/%s.json' % (dateBucket, hourBucket, city)
        fhand = open(outfile, 'x')
        fhand.write(r.text)
        fhand.close()
        time.sleep(5)
    except FileExistsError:
        pass
```

```python
import pandas as pd
import json
import os.path, time
from datetime import datetime

# parse scraped store list json file for a city
stores = pd.DataFrame()
for file in files:
    gmtime = time.gmtime(os.path.getctime(file))
    try:
        data = pd.read_json(file)['data'][0]
        df = pd.DataFrame(data)
        df['scrapeDate'] = str(gmtime[1]) + '/' + str(gmtime[2]) + '/' + str(gmtime[0])
        df['scrapeHour'] = gmtime[3]
        stores = pd.concat([stores, df])
    except KeyError:
        print(file)

# filter for relevant fields
fields = ['cityId', 'cityName', 'countryId', 'openStatus', 'storeId', 'storeName']
fields = ['storeId', 'cityName', 'openStatus', 'openTime', 'storeName', 'storeStar', 
          'serviceStar', 'envStar', 'tasteStar', 'speedStar', 'isOpenNet',
         'scrapeHour', 'scrapeDate']

# engineer new features: operating status
openStatus = {1:'正在营业', 12:'装修升级', 13:'迁址停业', 14:'暂停营业', 15:'试运 营', 5:'?', 11:'?'}
stores['operatingStatus'] = stores['openStatus'].apply(lambda x: openStatus[x])
fields.append('operatingStatus')

# function to return a city's tier
def cityTier(cityName, regionMark):
    tier1 = ['北京市', '上海市', '广州市', '深圳市']
    tier2 = ['合肥市', '重庆市', '福州市', '兰州市', '南宁市', '贵阳市', '海口市', '石家庄市', '哈尔滨市'
            , '郑州市', '武汉市', '长沙市', '呼和浩特市', '南京市', '南昌市', '长春市', '沈阳市', '银川市'
            , '西宁市', '西安市', '济南市', '太原市', '成都市', '天津市', '拉萨市', '乌鲁木齐市', '昆明市'
            , '杭州市', '青岛市', '厦门市', '宁波市', '大连市', '苏州市', '无锡市']
    if regionMark == 2:
        return 'International'
    elif (cityName in tier1) and (regionMark == 1):
        return 'Tier 1'
    elif (cityName in tier2) and (regionMark == 1):
        return 'Tier 2'
    else:
        return 'Tier 3'

# engineer new features: city tier
stores['tier'] = stores.apply(lambda x: cityTier(x['cityName'], x['regionMark']), axis=1)
fields.append('tier')

# engineer new features: max seat people
pattern = r'超过([0-9])人'
stores['maxSeatPeople'] = stores['dialogMessage'].str.extract(pattern)
fields.append('maxSeatPeople')

stores[fields].head()
#stores[fields].groupby('tier').nunique()
```
